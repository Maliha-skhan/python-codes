```python
countries = pd.read_csv("countries.csv")
```

```python
countries
```

```python
# Compare the population growth in the US and China
```

```python
us = countries[countries.country == "United States"]
```

```python
us
```

```python
china =  countries[countries.country == "China"]
```

```python
china
```

```python
plt.plot(us.year,us.population / 10**6)
plt.plot(china.year,china.population / 10**6)
plt.xlabel("year")
plt.ylabel("population in million")
plt.legend(["United States", "China"])
plt.show()
```

```python
us.population
```

```python
us.population / us.population.iloc[0]
```

```python
# Comparing the percentage growth per population
plt.plot(us.year,us.population * 100 / us.population.iloc[0])
plt.plot(china.year,china.population * 100 / china.population.iloc[0])
plt.xlabel("year")
plt.ylabel("percentage growth in population (first year = 100)")
plt.legend(["United States", "China"])
plt.savefig('test_plot')
plt.show()
```
